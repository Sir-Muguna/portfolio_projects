```python
import csv
import requests
from bs4 import BeautifulSoup

url = "https://www.redfin.com/city/30818/TX/Austin"

# Send an HTTP request to the URL and get the HTML content
response = requests.get(url)

# Check if the request was successful (status code 200 means success)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the property cards that contain the data
    property_cards = soup.find_all('div', class_='property-card')

    # Create a list to store the extracted data
    data_list = []

    # Loop through each property card and extract the desired information
    for property_card in property_cards:
        sale_type = property_card.find('span', class_='homecardIconLabel').text.strip()
        sold_date = property_card.find('span', class_='homecardSoldDate').text.strip()
        property_type = property_card.find('span', class_='property-type').text.strip()
        address = property_card.find('span', class_='address').text.strip()
        city = property_card.find('span', class_='citystatezip').text.strip().split(',')[0]
        state_or_province = property_card.find('span', class_='citystatezip').text.strip().split(',')[1].strip().split(' ')[0]
        zip_or_postal_code = property_card.find('span', class_='citystatezip').text.strip().split(',')[1].strip().split(' ')[1]
        price = property_card.find('span', class_='homecardPrice').text.strip().replace('$', '').replace(',', '')
        beds = property_card.find('span', class_='data-value', attrs={'data-rf-test-id': 'abp-beds'}).text.strip()
        baths = property_card.find('span', class_='data-value', attrs={'data-rf-test-id': 'abp-baths'}).text.strip()
        location = property_card.find('span', class_='homecardV2LocationText').text.strip()
        square_feet = property_card.find('span', class_='data-value', attrs={'data-rf-test-id': 'abp-sqFt'}).text.strip().replace(',', '')
        lot_size = property_card.find('span', class_='data-value', attrs={'data-rf-test-id': 'abp-lotSize'}).text.strip()
        year_built = property_card.find('span', class_='data-value', attrs={'data-rf-test-id': 'abp-yearBuilt'}).text.strip()
        days_on_market = property_card.find('span', class_='property-status').text.strip().split()[0]
        price_per_square_feet = property_card.find('span', class_='info-block', text=lambda t: '$/sqft' in t).text.strip().replace('$/sqft', '').replace(',', '')
        hoa_month = property_card.find('span', class_='info-block', text=lambda t: 'HOA/month' in t).text.strip().replace('HOA/month', '').replace('$', '').replace(',', '')
        status = property_card.find('span', class_='property-status').text.strip().split()[1]
        next_open_house_start_time = property_card.find('span', class_='avm-value').text.strip().split()[0]
        next_open_house_end_time = property_card.find('span', class_='avm-value').text.strip().split()[2]
        url = "https://www.redfin.com" + property_card.find('a', class_='slider-item')['href']
        source = property_card.find('span', class_='source-label').text.strip()
        mls = property_card.find('span', class_='mls-num').text.strip()
        favorite = property_card.find('span', class_='home-favorite').text.strip()
        interested = property_card.find('span', class_='home-interested').text.strip()
        latitude = property_card['lat']
        longitude = property_card['long']

        # Append the extracted data as a dictionary to the data_list
        data_list.append({
            "SALE TYPE": sale_type,
            "SOLD DATE": sold_date,
            "PROPERTY TYPE": property_type,
            "ADDRESS": address,
            "CITY": city,
            "STATE OR PROVINCE": state_or_province,
            "ZIP OR POSTAL CODE": zip_or_postal_code,
            "PRICE": price,
            "BEDS": beds,
            "BATHS": baths,
            "LOCATION": location,
            "SQUARE FEET": square_feet,
            "LOT SIZE": lot_size,
            "YEAR BUILT": year_built,
            "DAYS ON MARKET": days_on_market,
            "$/SQUARE FEET": price_per_square_feet,
            "HOA/MONTH": hoa_month,
            "STATUS": status,
            "NEXT OPEN HOUSE START TIME": next_open_house_start_time,
            "NEXT OPEN HOUSE END TIME": next_open_house_end_time,
            "URL": url,
            "SOURCE": source,
            "MLS#": mls,
            "FAVORITE": favorite,
            "INTERESTED": interested,
            "LATITUDE": latitude,
            "LONGITUDE": longitude,
        })

    # Define the CSV file path
    csv_file_path = "property_data.csv"

    # Write the data to a CSV file
    with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
        fieldnames = data_list[0].keys()
        writer = csv.DictWriter(file, fieldnames=fieldnames)

        # Write the header row
        writer.writeheader()

        # Write the data rows
        writer.writerows(data_list)

    print("Data has been successfully saved to:", csv_file_path)

else:
    print("Failed to fetch the page. Status code:", response.status_code)
```
