```python
# Load the data
# Import required packages
import pandas as pd

# File paths
text_data = "/root/projects/portfolio_projects/data_analytics/civic_technology_project/data/text_data.csv"
talent_survey = "/root/projects/portfolio_projects/data_analytics/civic_technology_project/data/talent_survey.csv"

# Load data
text_df = pd.read_csv(text_data)
survey_df = pd.read_csv(talent_survey)

# Preview the dataframes
print(text_df.head())
print(survey_df.head())
```

```python
# Define a Function to Check Inconsistencies in the dataframes
def check_data_quality(df, name="Dataset"):
    print(f"\n Data Quality Report for: {name}")
    print("-" * 50)
    print(" Shape:", df.shape)
    print("\n Columns & Types:")
    print(df.dtypes)
    print("\n Missing Values:")
    print(df.isnull().sum())
    print("\n Duplicate Rows:")
    print(df.duplicated().sum())
```

```python
# Run Checks on text_df dataframe
check_data_quality(text_df, name="Text Data")
```

```python
# Run Checks on survey_df dataframe
check_data_quality(survey_df, name="Talent Survey")
```

```python
# Data Cleaning & Processing in the survey_df
# Purpose: Standardize column names and handle missing values

# Standardize column names for better readability and analysis
new_columns = [
    'Token_ID',                     
    'Submission_Date',              
    'Gender',                      
    'Age_Group',                    
    'Institution_Type',             
    'Lecture_Satisfaction_Rating',  
    'Course_Relevance_To_Industry_Rating',  
    'Number_of_Industrial_Trainings_Required',  
    'Programming_Languages_In_Curriculum'  
]
# Apply the new column names
survey_df.columns = new_columns

# Convert date from format '2023-01-26T09:33:23Z' to simple date '2023-01-26'
survey_df['Submission_Date'] = pd.to_datetime(survey_df['Submission_Date']).dt.date

# Check missing values in the rating column
missing_count = survey_df['Course_Relevance_To_Industry_Rating'].isna().sum()
total_count = len(survey_df)
print(f"\nMissing Values Report:")
print(f"• Missing responses: {missing_count}/{total_count} ({missing_count/total_count:.1%})")

# Handling missing values: Preserve missingness as explicit "No Response" category
survey_df['Course_Relevance_To_Industry_Rating'] = survey_df['Course_Relevance_To_Industry_Rating'].fillna("No Response")

# Define output path and save cleaned data
output_dir = "/root/projects/portfolio_projects/data_analytics/civic_technology_project/data"
output_path = os.path.join(output_dir, "cleaned_talent_survey.csv")

# Save to CSV
survey_df.to_csv(output_path, index=False)


print("\nPost-Cleaning Value Counts:")
print(survey_df['Course_Relevance_To_Industry_Rating'].value_counts(dropna=False))

# Run Checks on survey_df dataframe
check_data_quality(survey_df, name="Talent Survey")
```
